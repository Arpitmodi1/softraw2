```python
# URL of the CSV file
url = ('https://raw.githubusercontent.com/daenuprobst/covid19-cases-switzerland/master/covid19_hospitalized_switzerland_openzh.csv')

# Downloading the file using wget
!wget -O dataset.csv $url

# Reading the CSV file into a pandas DataFrame
import pandas as pd
df = pd.read_csv('dataset.csv')

# Displaying the first few rows of the DataFrame
df.head()
```

```python
# 1. Read input files as data frames.
df_hosp = pd.read_csv("https://raw.githubusercontent.com/daenuprobst/covid19-cases-switzerland/master/covid19_hospitalized_switzerland_openzh.csv") # Read the correct file for df_hosp
df_demo = pd.read_csv("https://raw.githubusercontent.com/daenuprobst/covid19-cases-switzerland/master/demographics.csv")

# 2. Update the index of both DataFrames.
df_hosp.index = df_hosp["Date"]
df_demo.index = df_demo["Canton"]
df_demo.head()
df_hosp.head()
```

```python
# 3. Subset the number of hospitalized person data frame.
# Keep only columns "AG" -> "CH" (i.e. the first 27 columns).
df_hosp = df_hosp.loc[:, "AG":"CH"]
df_hosp.head()
```

```python
# 4. Extract population count per Canton
pop_by_canton = df_demo.loc[df_hosp.columns, "Population"]
```

```python
# 5. Compute hospitalized people relative to population size
df_hosp /= pop_by_canton / 10000
df_hosp.head()
```
